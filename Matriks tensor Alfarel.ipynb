```python
import numpy as np
```

```python
# 1. Buat matriks X berukuran 2x3
X = np.array([[10, 20, 30], [40, 50, 60]])
print("Matriks X:\n", X)
```

```python
# 2. Tampilkan ukuran dari matriks X
print("Ukuran matriks X:", X.shape)
```

```python
# 3. Tampilkan seluruh elemen dari kolom pertama (index 0)
print("Kolom pertama:", X[:, 0])
```

```python
# 4. Tampilkan seluruh elemen dari baris kedua (index 1)
print("Baris kedua:", X[1, :])
```
